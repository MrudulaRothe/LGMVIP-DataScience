## <i> Image to Pencil Sketch with Python - Mrudula Rothe</i>

```python
import cv2
```

```python
image = cv2.imread("image.jpg")
from IPython.display import Image
Image(filename='image.jpg')
```

```python
# Converting Original Image To Gray Scale

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('GRAY Image',gray)
cv2.waitKey(0)
```

```python
# Inverting the Gray Scale Image

invert = cv2.bitwise_not(gray)

cv2.imshow('Inverted Image',invert)
cv2.waitKey(0)
```

```python
# Next step is to blur the Image using Gaussian Function

blurred = cv2.GaussianBlur(invert, (21, 21), 0)
cv2.imshow('Blurred Image',blurred)
cv2.waitKey(0)
```

```python
inverted_blur = cv2.bitwise_not(blurred)
cv2.imshow('Inverted Blurred Image',inverted_blur)
cv2.waitKey(0)
```

```python

pencil_sketch = cv2.divide(gray, inverted_blur, scale=256.0)
cv2.imshow('Pencil Sketch',pencil_sketch)
cv2.waitKey(0)
```
